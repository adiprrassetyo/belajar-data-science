## Outlier

```python
df_num=df_train[numerical]
Q1=df_num.quantile(0.25)
Q3=df_num.quantile(0.75)
IQR=Q3-Q1
```

```python
low_limit=Q1-(IQR*1.5)
high_limit=Q3+(IQR*1.5)
```

```python
((df_train[numerical] < low_limit) | (df_train[numerical] > high_limit)).any()
```

```python
x=((df_train[numerical] < low_limit) | (df_train[numerical] > high_limit))
x['res']=x.any(axis=1)
x[x['res']==True].shape
```

```python
df_train=df_train[df_train.index.isin(x[x.res==False].index.to_list())]
df_train
```

## Model

```python
X=df_train.drop(['churn'],axis=1)
Y=df_train['churn']
X.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 123)
```

### Model Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model= LogisticRegression()
model_logreg = model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
y_pred_proba_logreg = model.predict_proba(X_test)[:,1]
log_reg = model.score(X_test, Y_test)
report_logreg = classification_report(Y_test, y_pred)
print(report_logreg)
```

```python
from sklearn.metrics import roc_auc_score

auc_logreg = roc_auc_score(Y_test,y_pred_proba_logreg)
print("AUC of ROC: %.3f%%" % (auc_logreg*100.0))
```

### Model Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model= RandomForestClassifier()
model_ranfor = model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
y_pred_proba_ranfor = model.predict_proba(X_test)[:,1]
ranfor = model.score(X_test, Y_test)
report_ranfor = classification_report(Y_test, y_pred)
print(report_ranfor)
```

```python
from sklearn.metrics import roc_auc_score

auc_ranfor = roc_auc_score(Y_test,y_pred_proba_ranfor)
print("AUC of ROC: %.3f%%" % (auc_ranfor*100.0))
```

### Model Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:,1]
dectre = model.score(X_test, Y_test)
report_dectre = classification_report(Y_test, y_pred)
print(report_dectre)
```

```python
from sklearn.metrics import roc_auc_score

auc_dectre = roc_auc_score(Y_test,y_pred_proba)
print("AUC of ROC: %.3f%%" % (auc_dectre*100.0))
```
